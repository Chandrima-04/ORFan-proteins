```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(open("orfan_metrics/summary_all_20111209.html"), 'html.parser')
```

```python
header = [str(x.string) for x in soup.findAll("thead")[0].findAll("tr")[1].findAll("th")][:-4]
rows = [ tuple(str(element.contents[0]) for element in row.findAll("td")[:-4]) for row in soup.findAll("tr") ]
```

```python
df = pd.DataFrame.from_records(rows[2:], columns=header)
df.index = df["Cluster"]
df = df.sort_values(by ="composite score",ascending=False)
```

```python
print(df.shape)
df.head()
```

```python
#Add cluster sample source info to the table
table3 = pd.read_csv("orfan_metrics/Messina_Table3.csv",sep=";")
table3 = table3[["cluster_id","sample_origin"]]
table3.head()
```

```python
df = df.merge(table3,left_on="Cluster",right_on="cluster_id")
```

```python
df.to_csv("1_out/cluster_stats.csv",index=False)
```
