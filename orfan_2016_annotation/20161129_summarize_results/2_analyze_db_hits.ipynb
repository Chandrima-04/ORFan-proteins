```python
import pandas as pd
import re
import itertools
import matplotlib_venn
import numpy as np
```

# 1. Load blast hits , hmmer hits and cluster stats

```python
#Load blast best hits
filt_blast_hits = pd.read_csv("1_out/filtered_blast_best_hits.csv")
print(filt_blast_hits.shape)
#Load hmmer best hits
filt_hmmer_hits = pd.read_csv("1_out/filtered_hmmer_best_hits.csv")
print(filt_hmmer_hits.shape)

all_filt_hits = pd.concat([filt_blast_hits,filt_hmmer_hits],axis=0)
#Load cluster stats
clusters = pd.read_csv("1_out/cluster_stats.csv")
clusters["sample_origin"] = clusters["sample_origin"].apply(lambda x:x.replace(",",";"))
```

```python
print(all_filt_hits.shape)
all_filt_hits.head()
```

```python
print(clusters.shape)
clusters.head()
```

# 2. Summary of db hits per cluster

```python
tmp = [(cluster,hits.shape[0],";".join(hits.db)) for cluster,hits in ( all_filt_hits[["cluster","db"]].drop_duplicates().groupby("cluster") )]
hits_x_cluster = pd.DataFrame.from_records(tmp,columns=["cluster","db_hit_count","dbs"])
print("Cluster families with no annotation: {}".format(clusters.shape[0]-hits_x_cluster.shape[0]))
hits_x_cluster = hits_x_cluster.merge(clusters,left_on="cluster",right_on="Cluster",how="right")[["Cluster","db_hit_count","dbs","sample_origin","composite score","P"]]
print(hits_x_cluster.shape)
hits_x_cluster = hits_x_cluster.sort_values(by="composite score",ascending=False)
hits_x_cluster.head()
```

```python
hits_x_cluster.to_csv("2_out/dbhits_x_cluster.csv",index=False,quote="\"")
```

# 3. Summary of db hits grouped by sample origin

### Hits per db per sample_origin

```python
hits_x_sample_origin = {}
for db in all_filt_hits.db.unique():
    hits_x_sample_origin[db] = hits_x_cluster.groupby("sample_origin").apply(lambda df: sum([(db in x.split(";")) if type(x) is str else False for x in df["dbs"]]))
hits_x_sample_origin = pd.concat(hits_x_sample_origin,axis=1)
hits_x_sample_origin
```

### Calculate hit count to metahit = (metahit_pep or metahit_cds ), and env = ( env nr or env nt ) 

```python
hits_x_sample_origin_sum = {}
for db in ["env_","metahit_","metahit_2014"]:
    hits_x_sample_origin_sum[db] = hits_x_cluster.groupby("sample_origin").apply(lambda df: sum([(db in x) if type(x) is str else False for x in df["dbs"]]))
hits_x_sample_origin_sum = pd.concat(hits_x_sample_origin_sum,axis=1)
hits_x_sample_origin_sum
```

```python
table_2 = pd.concat([hits_x_cluster.groupby("sample_origin").count()["Cluster"],
            hits_x_sample_origin,
            hits_x_sample_origin_sum],axis=1)
table_2.to_csv("2_out/dbhits_x_sample_origin.csv")
table_2
```

# 4. Analyze NCBI Environmental hits

```python
#Extract all env hits
env_hits = all_filt_hits[["cluster","db","s_description"]][all_filt_hits.db.apply(lambda x: "env_" in x)].sort_values(by="cluster")
#Parse description to extract the source metagenome
env_hits["env_source"] = env_hits.s_description.apply(lambda x: x.split("metagenome")[0])
env_hits["env_source"] = env_hits["env_source"].apply(lambda x: x.split("[")[1] if "[" in x else x   )
#Annotate clusters with sample origin
env_hits = env_hits.merge(clusters[["Cluster","sample_origin","composite score"]],left_on="cluster",right_on="Cluster")
env_hits = env_hits.drop(["s_description","Cluster"],axis=1)
env_hits.index = env_hits["cluster"]
#Remove duplicates when blastp and hmmsearch hits are the same
env_hits = env_hits.drop_duplicates()
print(env_hits.shape)
env_hits.head()
```

## 4.1 Environmental hits per cluster

```python
tmp = env_hits.groupby("cluster").apply( lambda df: ";".join( "{}({})".format(env_src,"n" if db == "env_nt" else "p") for _,db,env_src in df[["db","env_source"]].drop_duplicates().to_records()) )
env_hits_x_clust = pd.merge(clusters[["Cluster","composite score","sample_origin"]], 
                             pd.DataFrame(tmp,columns=["env_hits"]),
                             left_on="Cluster",right_index=True,how="left").sort_values(by="composite score",ascending=False)
env_hits_x_clust.to_csv("2_out/env_hits_x_cluster.csv",index=False)
env_hits_x_clust.head()
```

## 4.2 Environmental hits per sample origin

```python
env_hits_x_sample_type = pd.DataFrame(env_hits.groupby("sample_origin").apply(lambda df: "; ".join("{}({})".format(k,v) for k,v in df.groupby("env_source").count()["cluster"].to_dict().items())))
env_hits_x_sample_type.columns = ["env_hits"]
env_hits_x_sample_type.to_csv("2_out/env_hits_per_sample_origin.csv")
env_hits_x_sample_type
```

# 5. Summarize MetaHIT hits

```python
metahit_hits = all_filt_hits[["cluster","db"]][all_filt_hits.db.apply(lambda x: "metahit_" in x)].sort_values(by="cluster")
metahit_x_cluster = metahit_hits.drop_duplicates().groupby("cluster").apply(lambda df: ";".join(sorted(df.db)))
metahit_x_cluster= pd.DataFrame(metahit_x_cluster.apply(lambda x: x.replace("metahit_","").replace("cds","n").replace("pep","p")),columns=["MetaHIT"])
metahit_x_cluster = pd.merge(clusters[["Cluster","composite score","sample_origin"]], 
                             metahit_x_cluster,
                             left_on="Cluster",right_index=True,how="left")
metahit_x_cluster.sort_values(by="composite score",ascending=False).to_csv("2_out/metahit_hits_x_cluster.csv",index=False)
metahit_x_cluster.head()
```
